# Exercice 2

```python
class Curve:
    def __init__(self,start,stop,nbr_points=5432):
        self.start=start
        self.stop=stop
        self.nbr_points=nbr_points
        
    def plot_curve(self):
        x_val=np.linspace(self.start,self.stop,1000)
        y_val=x_val**5+0.5
        plt.plot(x_val,y_val,color='red',label='$y=x^5$')
        plt.xlabel('x')
        plt.ylabel('y')
        plt.legend()
        plt.grid(True)
        plt.title("Courbe d'quation y=x^5")
        
    def superieur(self,x,y):
        return y>x**5+0.5
    
    def generation_points(self):
        random_x=np.random.uniform(self.start,self.stop,self.nbr_points)
        random_y=np.random.uniform(self.start,self.stop,self.nbr_points)
        surface_sup=0
        surface_inf=0
        n=1000
        nbx=1/n
        for i in range(n):
            x_left=i*nbx
            x_right=(i+1)*nbx
            y_left=x_left**5+0.5
            y_right=x_right**5+0.5
            for x,y in zip(random_x,random_y):
                if y>x**5+0.5:
                    plt.scatter(x,y,color='blue',marker='x')
                    surface_sup+=y-(((x-1)**5)+0.5)
                else:
                    plt.scatter(x,y,color='green',marker='o')
                    surface_inf+=(((x-1)**5)+0.5)-y
        print(f"Surface bleu:{surface_sup*nbx}")
        print(f"Surface vert:{surface_inf*nbx}")
    
```

```python
curve=Curve(0,1)
curve.plot_curve()
curve.generation_points()
plt.show()
```
